```python
## 2. Null values. Let's see at column lvek the number of nulls.

null_cols = df2.isnull().sum()
print(null_cols[null_cols > 0])
```

```python
## Lets's try to uniform the ACtivity column

l1=[str(e).lower().find("surf") + str(e).lower().find("boarding") for e in df3['Activity']]
l2=[str(e).lower().find("fishing") for e in df3['Activity']]
l3=[str(e).lower().find("swim")+str(e).lower().find("wading")+str(e).lower().find("walking")+str(e).lower().find("bathing")+str(e).lower().find("treading") for e in df3['Activity']]
l4=[str(e).lower().find("diving")+str(e).lower().find("snorkel") for e in df3['Activity']]

df31=pd.concat([pd.Series(l1),pd.Series(l2),pd.Series(l3),pd.Series(l4)],axis=1)

df31[0] = np.where(df31[0].astype(int)>-2,1,0)
df31[1] = np.where(df31[1].astype(int)>-1,1,0)
df31[2] = np.where(df31[2].astype(int)>-5,1,0)
df31[3] = np.where(df31[3].astype(int)>-2,1,0)

df31[4] = df31[0]*1000+df31[1]*100+df31[2]*10+df31[3]

a=("surfing","fishing","swimming","diving","other")

activity2=[]
for e in df31[4]:
    if e>=1000:
        activity2.append(a[0])
    elif e>=100:
        activity2.append(a[1])
    elif e>=10:
        activity2.append(a[2])
    elif e>=1:
        activity2.append(a[3])
    else:
        activity2.append(a[4])
'''
df31[4] = np.where(df31[4].astype(int)>=1000,str("surfing"),str("other"))
df31[4] = np.where(df31[4].astype(int)>=100,"fishing","other")
df31[4] = np.where(df31[4].astype(int)>=10,"swimming","other")
df31[4] = np.where(df31[4].astype(int)>=1,"diving","other")
'''

activity2=pd.DataFrame(activity2)
activity2_c=activity2[0].value_counts()

display (activity2_c)

## np.where(df['First Season'] > 1990, 1, df['First Season'])
## df.First Season.loc[(df['First Season'] > 1990)] = 1
## x = re.search(r"\bS\w+", str)
## print(x.string) 
```
