```python
## 1. Let's look at what data we have (columns and type) and how many data elements (shape)
print(df.shape)
print(df.dtypes)
```

```python
## There are three colums with similar names (Case Number{})
## Let's see if the elements are duplicated by generating a new data frame only with duplicated elements between columns.

dup1 = df[(df['Case Number.1'])==(df['Case Number'])]
dup2 = df[(df['Case Number.2'])==(df['Case Number'])]

print(dup1.shape, dup2.shape)
```

```python
## Out of the total 5992 rows in the data 5979 and 5990 are duplicated for 'Case Number.1' and 'Case Number.2'
## Removing columns 

df21 = df.drop('Case Number.1',1)
df2 = df21.drop('Case Number.2',1)
print(df2.shape)
```

```python
## 2. Null values. Let's see at column lvek the number of nulls.

null_cols = df2.isnull().sum()
print(null_cols[null_cols > 0])
```
